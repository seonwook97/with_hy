```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')
```

```python
def total_df(merged, name_str):
  temp = pd.read_csv(f'/content/drive/MyDrive/final_project/{name_str}.csv',index_col=0)
  temp.drop_duplicates(subset='headline',inplace=True)
  return pd.concat([merged,temp])

# 5개년 데이터와 10개년 데이터 크롤링이 중복되어있어 10년 데이터에 맞게 전부 read 후 concat 진행하기
names = ['삼성전자','sk','lg화학','naver','현대자동차','celltrion','셀트리온','posco','POSCO','카카오','kakao','KB금융','kb','신한지주']
df = pd.DataFrame()
for name in names:
  df = total_df(df, name)

df 
```

```python
df.reset_index(inplace=True)
del df['index']
df
```
